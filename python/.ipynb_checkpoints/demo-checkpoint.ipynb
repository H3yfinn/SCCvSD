```python
import pyflann
import scipy.io as sio
import numpy as np
import cv2 as cv
import time

from util.synthetic_util import SyntheticUtil
from util.iou_util import IouUtil
from util.projective_camera import ProjectiveCamera

import argparse

#parser = argparse.ArgumentParser()
#parser.add_argument('--feature-type', required=True, type=str, default='deep', help='deep or HoG')
#parser.add_argument('--query-index', required=True, type=int, default=0, help='[0, 186)')

#args = parser.parse_args()
feature_type = 'deep'# args.feature_type
#assert feature_type == 'deep' or feature_type == 'HoG'
query_index = 0#args.query_index
#assert 0 <= query_index < 186
#
"""
Estimate an homogrpahy using edge images 
"""

# Step 1: load data
# database
if feature_type == 'deep':
    data = sio.loadmat('../data/features/database_camera_feature.mat')
    database_features = data['features']
    database_cameras = data['cameras']
else:
    data = sio.loadmat('../data/features/database_camera_feature_HoG.mat')
    database_features = data['features']
    database_cameras = data['cameras']#finn so i think these have recorded all possible camera permutations

# testing edge image from two-GAN #finn why cant this be a football broadcast image isntead of edge image?
if feature_type == 'deep':
    data = sio.loadmat('../data/features/testset_feature.mat')#finn how to generate these
    edge_map = data['edge_map']#finn how to generate these
    test_features = data['features']#finn how to generate these
    test_features = np.transpose(test_features)#finn how to generate these
else:
    data = sio.loadmat('../data/features/testset_feature_HoG.mat')#finn hog is the baseline
    edge_map = data['edge_map']
    test_features = data['features']

# World Cup soccer template
data = sio.loadmat('../data/worldcup2014.mat')
model_points = data['points']
model_line_index = data['line_segment_index']

template_h = 74  # yard, soccer template
template_w = 115


# ground truth homography # finn why has this been loaded?to check on accuracy?
data = sio.loadmat('../data/UoT_soccer/test.mat')
annotation = data['annotation']
gt_h = annotation[0][query_index][1]  # ground truth


state_time = time.time()
# Step 2: retrieve a camera using deep features #finn i think this finds the nearest neighbour for the test features
flann = pyflann.FLANN()#fast library approx nearest neighbours
result, _ = flann.nn(database_features, test_features[query_index], 1, algorithm="kdtree", trees=8, checks=64)
retrieved_index = result[0]


"""
Retrieval camera: get the nearest-neighbor camera from database #finn uses the output from FLANN above. literally a dict search now
"""
retrieved_camera_data = database_cameras[retrieved_index]

u, v, fl = retrieved_camera_data[0:3]
rod_rot = retrieved_camera_data[3:6]
cc = retrieved_camera_data[6:9]

retrieved_camera = ProjectiveCamera(fl, u, v, cc, rod_rot)

retrieved_h = IouUtil.template_to_image_homography_uot(retrieved_camera, template_h, template_w)
#template h , w mean height and width of image template. i think referring to the topdwon view as well. 

iou_1 = IouUtil.iou_on_template_uot(gt_h, retrieved_h)
print('retrieved homogrpahy IoU {:.3f}'.format(iou_1))
```

```python
def find_transform(im_src, im_dst):
    warp = np.eye(3, dtype=np.float32)
    criteria = (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, 50, 0.001)
    #print(criteria, warp, cv.MOTION_HOMOGRAPHY)
    try:
        _, warp = cv.findTransformECC(im_src, im_dst, warp, cv.MOTION_HOMOGRAPHY, criteria, None,1)#finn
    except:
        print('Warning: find transform failed. Set warp as identity')
    return warp
```
